The core concepts of Object-Oriented Programming (OOP) in Python, which is a programming paradigm that uses objects and classes to structure software programs. Understanding these concepts is crucial for writing efficient, reusable, and organized code. Here are the foundational OOP concepts explained:

### 1. Classes and Objects
- **Classes:** A class is a blueprint for creating objects. It defines a set of attributes and methods that the created objects (instances) will have. In Python, you define a class using the `class` keyword.
- **Objects:** An object is an instance of a class. It has the class-defined attributes and methods. Objects are created by calling the class like a function.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return "Woof!"

# Creating an object of the Dog class
my_dog = Dog("Buddy", 3)
print(my_dog.bark())  # Outputs: Woof!
```

### 2. Encapsulation
Encapsulation is the mechanism of hiding the internal state of an object and requiring all interaction to be performed through an object's methods. This provides a protective barrier that prevents the data from being directly accessed or modified from outside the object.

```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return "Deposit successful"

    def get_balance(self):
        return self.__balance
```

### 3. Inheritance
Inheritance allows a class (child class) to inherit attributes and methods from another class (parent class). This promotes code reuse and a hierarchical organization of classes.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"
```

### 4. Polymorphism
Polymorphism allows methods to do different things based on the object it is acting upon. This is closely related to inheritance and can be achieved through method overriding where a child class redefines a method from its parent class.

```python
# Using the Animal, Dog, and Cat classes from the Inheritance example

dog = Dog("Buddy")
cat = Cat("Whiskers")

for animal in (dog, cat):
    print(animal.speak())
```

### 5. Abstraction
Abstraction involves creating simple interfaces while hiding the underlying complexity. Abstract classes and methods define what types of operations an object can perform, without specifying how these operations will be carried out.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
```

````python
class DataPreprocessor:
    def __init__(self, data):
        self.data = data

    def clean_data(self):
        self.data = [row for row in self.data if row['size'] >= 400 and row['bedrooms'] <= 10]
    
    def scale_features(self):
        sizes = [row['size'] for row in self.data]
        min_size, max_size = min(sizes), max(sizes)
        for row in self.data:
            row['size'] = (row['size'] - min_size) / (max_size - min_size)
    
    def encode_categorical(self):
        neighborhoods = list(set(row['neighborhood'] for row in self.data))
        neighborhood_map = {n: i for i, n in enumerate(neighborhoods)}
        for row in self.data:
            row['neighborhood'] = neighborhood_map[row['neighborhood']]

# Example dataset
data = [
    {'size': 500, 'bedrooms': 2, 'neighborhood': 'Downtown'},
    {'size': 750, 'bedrooms': 3, 'neighborhood': 'Suburb'},
    {'size': 200, 'bedrooms': 1, 'neighborhood': 'Downtown'},  # This row should be removed by clean_data
    {'size': 850, 'bedrooms': 2, 'neighborhood': 'Downtown'}
]

preprocessor = DataPreprocessor(data)
preprocessor.clean_data()
preprocessor.scale_features()
preprocessor.encode_categorical()

# Print preprocessed data
print(preprocessor.data)
````

# Reading and Writing in Python Files 

File operations in Python are essential for reading from and writing to files on your filesystem. Python provides a built-in function `open()` that is used to open a file in either read (`'r'`), write (`'w'`), append (`'a'`), or read-write (`'r+'`) mode.

When you open a file using the `open()` function, it returns a file object, which provides methods and attributes to perform various operations on the file such as reading from or writing to it. It's crucial to close the file after you're done with it to free up system resources. Python's `with` statement helps manage this by automatically closing the file once you leave the block.

### Basic File Operations
- **Reading from a file:** You can read the entire file as a string or iterate through lines in a file.
- **Writing to a file:** You can write or append content to a file. If the file does not exist, Python will create it for you in write or append mode.
- **Closing a file:** It's important to close your file to ensure that data is properly saved and resources are freed. Using `with` automatically takes care of this.

### Example: Reading from and Writing to a File

#### Writing to a File
Let's create a text file and write some lines to it.

```python
with open('example.txt', 'w') as file:
    file.write("Hello, world!\n")
    file.write("This is an example of writing to a file in Python.\n")
```

This code block opens (and creates if not existing) `example.txt` in write mode (`'w'`). The `with` statement ensures the file is properly closed after the block is executed. Two lines of text are written to the file using the `write()` method.

#### Reading from a File
Now, let's read the content we just wrote in the file and print it.

```python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

This opens `example.txt` in read mode (`'r'`), reads the content into the variable `content` using the `read()` method, and prints the content. Again, the `with` statement ensures that the file is closed after the block of code is executed.

#### Appending to a File
If you want to add more content to the end of the file without overwriting the existing content, you can open the file in append mode (`'a'`).

```python
with open('example.txt', 'a') as file:
    file.write("Appending a new line to the file.\n")
```

After executing this block, `example.txt` will contain the new line in addition to the previously written content.

### Key Points
- **Modes:** `'r'` for reading, `'w'` for writing (overwrites existing file), `'a'` for appending, and `'r+'` for reading and writing.
- **Automatic Closure:** The `with` statement handles opening and closing files, which is the recommended practice for file operations in Python.
- **Reading Methods:** `read()` reads the entire file, `readline()` reads the next line, and `readlines()` reads all the lines into a list.

# Exception Handling 

Exception handling in Python is crucial for managing errors that occur during the execution of a program. It allows you to gracefully handle error conditions, preventing the program from crashing and, if necessary, providing the user or developer with useful information about the problem.

Python uses the `try`, `except`, `else`, and `finally` blocks to handle exceptions.

- **`try` Block:** You write the code that might throw an exception inside this block.
- **`except` Block:** This block catches and handles the exception. You can specify the type of exception to catch, allowing different exceptions to be handled differently.
- **`else` Block:** If no exceptions were raised in the `try` block, the `else` block is executed.
- **`finally` Block:** This block is executed no matter what, and is typically used for cleaning up resources, like closing files or releasing external resources.

### Example: Handling a Division Operation

Let's create a simple example where we handle an exception that might occur during a division operation, such as dividing by zero.

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: You can't divide by zero!")
    else:
        print(f"The result is {result}")
    finally:
        print("Executing finally clause")

# Example usage
divide(10, 2)
divide(10, 0)
```

#### Explanation:

1. **First Call (`divide(10, 2)`):**
   - The `try` block executes without any exceptions since `y` is not zero.
   - The `else` block is executed, printing the division result.
   - The `finally` block is executed, printing the message about executing the finally clause.

2. **Second Call (`divide(10, 0)`):**
   - The `try` block raises a `ZeroDivisionError` because `y` is zero.
   - The `except ZeroDivisionError` block catches this exception and prints a message about not being able to divide by zero.
   - The `finally` block is executed, printing the message about executing the finally clause.

### Key Points:
- **Specific Exceptions:** You can catch specific exceptions to handle different error conditions in different ways. For example, catching a `FileNotFoundError` when attempting to open a file that doesn't exist.
- **Multiple `except` Blocks:** You can have multiple `except` blocks to handle different types of exceptions.
- **`else` Block:** It's a good practice to use the `else` block for code that should run if no exceptions were thrown in the `try` block, especially if the code in the `else` block might throw new exceptions that you don't want to catch in the same `try` block.
- **`finally` Block:** Use this for clean-up actions that must be executed under all circumstances, such as closing a file or releasing a network resource.

This structured approach to exception handling makes your Python programs more robust and error-tolerant.

Exercise 13: Database Connection Handling

Create a class for connecting to a database and handle exceptions related to database connection errors.

````python
import sqlite3

class DatabaseConnector:
    def __init__(self, database_file):
        self.database_file = database_file
    
    def connect(self):
        try:
            connection = sqlite3.connect(self.database_file)
            return connection
        except sqlite3.Error as e:
            print(f"Database connection error: {e}")
    
    def execute_query(self, query):
        try:
            connection = self.connect()
            cursor = connection.cursor()
            cursor.execute(query)
            return cursor.fetchall()
        except sqlite3.Error as e:
            print(f"Database error: {e}")
        finally:
            if connection:
                connection.close()

# Example usage:
db_connector = DatabaseConnector("mydb.db")
query = "SELECT * FROM mytable"
results = db_connector.execute_query(query)
````

Step-by-Step Explanation of Main Function
- Initialize the Library: A Library object is created, specifying the file path for storing the books data (books.json). This will load existing books from the file into the library's books list.
- Display Menu: A while loop presents a menu to the user with options to add, update, remove, list books, or exit.
- Handle User Input: Depending on the user's choice, the script calls the corresponding method in the Library object. It collects necessary information from the user through input prompts.
- Perform Actions: Each action modifies the library's books list or displays information. Changes to the book collection are persisted to the file through the save_books method.

````python
import json

class Book:
    def __init__(self, title, author, isbn, quantity):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.quantity = quantity

    def __repr__(self):
        return f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}, Quantity: {self.quantity}"

class Library:
    def __init__(self, filepath):
        self.books = []
        self.filepath = filepath
        self.load_books()

    def add_book(self, book):
        self.books.append(book)
        self.save_books()

    def update_book(self, isbn, title=None, author=None, quantity=None):
        for book in self.books:
            if book.isbn == isbn:
                if title:
                    book.title = title
                if author:
                    book.author = author
                if quantity is not None:
                    book.quantity = quantity
                self.save_books()
                return True
        return False

    def remove_book(self, isbn):
        for i, book in enumerate(self.books):
            if book.isbn == isbn:
                del self.books[i]
                self.save_books()
                return True
        return False

    def list_books(self):
        for book in self.books:
            print(book)

    def load_books(self):
        try:
            with open(self.filepath, 'r') as file:
                books_data = json.load(file)
                self.books = [Book(**book) for book in books_data]
        except FileNotFoundError:
            print("Books file not found. Starting with an empty library.")
        except json.JSONDecodeError:
            print("Error reading the books file. Data might be corrupted.")

    def save_books(self):
        with open(self.filepath, 'w') as file:
            json.dump([book.__dict__ for book in self.books], file)

def main():
    library = Library('books.json')
    
    while True:
        print("\nLibrary Management System")
        print("1. Add Book")
        print("2. Update Book")
        print("3. Remove Book")
        print("4. List Books")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            title = input("Enter book title: ")
            author = input("Enter author name: ")
            isbn = input("Enter ISBN: ")
            quantity = int(input("Enter quantity: "))
            library.add_book(Book(title, author, isbn, quantity))
        elif choice == '2':
            isbn = input("Enter ISBN of the book to update: ")
            title = input("Enter new title (leave blank to skip): ")
            author = input("Enter new author (leave blank to skip): ")
            quantity = input("Enter new quantity (leave blank to skip): ")
            quantity = int(quantity) if quantity else None
            if not library.update_book(isbn, title, author, quantity):
                print("Book not found.")
        elif choice == '3':
            isbn = input("Enter ISBN of the book to remove: ")
            if not library.remove_book(isbn):
                print("Book not found.")
        elif choice == '4':
            library.list_books()
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
````
