````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Data Collection
# URLs for the confirmed, recovered, and deaths datasets
url_confirmed = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
url_recovered = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
url_deaths = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"

# Reading data into DataFrames
df_confirmed = pd.read_csv(url_confirmed)
df_recovered = pd.read_csv(url_recovered)
df_deaths = pd.read_csv(url_deaths)
df = pd.read_csv(url_confirmed)
````

````python
# Task 2: Data Preparation & Cleaning
# Convert date columns to datetime
df.columns = pd.to_datetime(df.columns, errors='ignore')
# Handle missing values (if any)

# Task 3: Data Manipulation
# Aggregate global cases by date
global_cases = df.drop(['Lat', 'Long', 'Province/State', 'Country/Region'], axis=1).sum()

# Reshape for easier analysis (optional detailed step)

# Task 4: Exploratory Data Analysis
# Total and new daily cases calculation
global_daily_cases = global_cases.diff().fillna(global_cases.iloc[0]) # Calculate new daily cases
total_global_cases = global_cases.iloc[-1] # Total confirmed cases till the latest date

# Identify the top 5 countries with the highest total confirmed cases
top_5_countries = df.groupby('Country/Region').sum().iloc[:, -1].nlargest(5)

# Task 5: Visualization
# Line plot for global cases
plt.figure(figsize=(10, 6))
global_cases.plot()
plt.title('Global Confirmed Cases Over Time')
plt.ylabel('Total Confirmed Cases')
plt.xlabel('Date')
plt.show()

# Bar chart for top 5 affected countries
plt.figure(figsize=(10, 6))
top_5_countries.plot(kind='bar')
plt.title('Top 5 Countries with Highest Total Confirmed Cases')
plt.ylabel('Total Confirmed Cases')
plt.xlabel('Country')
plt.show()
````

````python
# Heatmap or line plot comparing new daily cases trend in top 5 countries
top_5_countries_names = top_5_countries.index
top_5_countries_data = df[df['Country/Region'].isin(top_5_countries_names)].groupby('Country/Region').sum().drop(['Lat', 'Long'], axis=1)
plt.figure(figsize=(12, 8))
sns.lineplot(data=top_5_countries_data.T, dashes=False)
plt.title('New Daily Cases Trend in Top 5 Affected Countries')
plt.ylabel('New Daily Cases')
plt.xlabel('Date')
plt.legend(top_5_countries_names)
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Data Loading and Cleaning
gtd_df = pd.read_csv('globalterrorismdb.csv', encoding='ISO-8859-1')  # Adjust the file name/path as necessary

# Display basic information about the dataset
print(gtd_df.info())

# Handling missing values
# Checking for missing values in each column
missing_values = gtd_df.isnull().sum()
print("Missing Values per Column:")
print(missing_values)

# Dropping irrelevant columns
columns_to_drop = ['eventid', 'approxdate', 'extended', 'resolution', 'country', 'region', 'provstate',
                   'city', 'latitude', 'longitude', 'specificity', 'vicinity', 'location', 'summary', 'doubtterr',
                   'alternative', 'alternative_txt', 'multiple', 'attacktype2', 'attacktype2_txt',
                   'attacktype3', 'attacktype3_txt', 'targsubtype1', 'targsubtype1_txt', 'targtype2', 'targtype2_txt',
                   'targsubtype2', 'targsubtype2_txt', 'corp2', 'target2', 'natlty2', 'natlty2_txt', 'targtype3',
                   'targtype3_txt', 'targsubtype3', 'targsubtype3_txt', 'corp3', 'target3', 'natlty3', 'natlty3_txt',
                   'gsubname', 'gname2', 'gsubname2', 'gname3', 'gsubname3', 'motive', 'guncertain1', 'guncertain2',
                   'guncertain3', 'nperps', 'nperpcap', 'claimed', 'claimmode', 'claimmode_txt', 'claim2',
                   'claimmode2', 'claimmode2_txt', 'claim3', 'claimmode3', 'claimmode3_txt', 'compclaim',
                   'weaptype2', 'weaptype2_txt', 'weapsubtype2', 'weapsubtype2_txt', 'weaptype3', 'weaptype3_txt',
                   'weapsubtype3', 'weapsubtype3_txt', 'weaptype4', 'weaptype4_txt', 'weapsubtype4', 'weapsubtype4_txt',
                   'nkillus', 'nkillter', 'nwoundus', 'nwoundte', 'property', 'propextent', 'propextent_txt',
                   'propvalue', 'propcomment', 'ishostkid', 'nhostkid', 'nhostkidus', 'nhours', 'ndays', 'divert',
                   'kidhijcountry', 'ransom', 'ransomamt', 'ransomamtus', 'ransompaid', 'ransompaidus', 'ransomnote',
                   'hostkidoutcome', 'hostkidoutcome_txt', 'nreleased', 'addnotes', 'scite1', 'scite2', 'scite3',
                   'related']
gtd_df.drop(columns=columns_to_drop, inplace=True)

# Handling missing values in remaining columns
# For simplicity, let's drop rows with missing values for now
gtd_df.dropna(inplace=True)

# Converting data types
gtd_df['iyear'] = pd.to_datetime(gtd_df['iyear'], format='%Y')

# Task 2: EDA - Number of Attacks Per Year
attacks_per_year = gtd_df.groupby(gtd_df['iyear'].dt.year).size()
plt.figure(figsize=(15, 6))
attacks_per_year.plot(kind='line')
plt.title('Global Terrorism Attacks Per Year')
plt.ylabel('Number of Attacks')
plt.xlabel('Year')
plt.show()

# Task 3: Geographical Analysis
# Analyzing distribution of attacks by country and region

# Grouping data by country
attacks_by_country = gtd_df['country_txt'].value_counts().head(10)

# Visualizing the top 10 countries with the most attacks
plt.figure(figsize=(10, 6))
sns.barplot(x=attacks_by_country.values, y=attacks_by_country.index, palette='viridis')
plt.title('Top 10 Countries with the Most Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()

# Task 4: Attack Types and Targets
# Investigating the most common types of attacks and targets

# Grouping data by attack type
attack_types = gtd_df['attacktype1_txt'].value_counts()

# Visualizing the distribution of attack types
plt.figure(figsize=(10, 6))
sns.barplot(x=attack_types.values, y=attack_types.index, palette='magma')
plt.title('Distribution of Attack Types')
plt.xlabel('Number of Attacks')
plt.ylabel('Attack Type')
plt.show()

# Grouping data by target type
target_types = gtd_df['targtype1_txt'].value_counts()

# Visualizing the distribution of target types
plt.figure(figsize=(10, 6))
sns.barplot(x=target_types.values, y=target_types.index, palette='rocket')
plt.title('Distribution of Target Types')
plt.xlabel('Number of Attacks')
plt.ylabel('Target Type')
plt.show()

# Task 5: Casualty Analysis
# Calculating and visualizing the number of killings and injuries resulting from terrorist attacks over time

# Creating a new column for total casualties
gtd_df['total_casualties'] = gtd_df['nkill'] + gtd_df['nwound']

# Grouping data by year and calculating total casualties
casualties_per_year = gtd_df.groupby(gtd_df['iyear'].dt.year)['total_casualties'].sum()

# Visualizing the trend of casualties over time
plt.figure(figsize=(15, 6))
casualties_per_year.plot(kind='line')
plt.title('Total Casualties from Terrorism Over Time')
plt.ylabel('Number of Casualties')
plt.xlabel('Year')
plt.show()

# Task 6: Perpetrator Group Analysis
# Identifying the most active terrorist groups based on the number of attacks

# Grouping data by terrorist group and counting the number of attacks
terrorist_groups = gtd_df['gname'].value_counts().head(10)

# Visualizing the top 10 most active terrorist groups
plt.figure(figsize=(10, 6))
sns.barplot(x=terrorist_groups.values, y=terrorist_groups.index, palette='coolwarm')
plt.title('Top 10 Most Active Terrorist Groups')
plt.xlabel('Number of Attacks')
plt.ylabel('Terrorist Group')
plt.show()

# Further analysis on preferred methods, targets, etc. can be conducted as well

# Task 7: Advanced Analysis (Optional)
# Time-series analysis and text analysis can be performed here
````
