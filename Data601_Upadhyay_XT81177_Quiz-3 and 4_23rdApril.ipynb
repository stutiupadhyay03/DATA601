````python
import pandas as pd
import os

def load_and_concatenate_sales_data(directory):
    """
    Concatenate multiple sales CSV files from a specified directory into a single DataFrame.

    Parameters:
        directory (str): The directory path containing sales CSV files.

    Returns:
        pandas.DataFrame: Concatenated DataFrame containing sales data.
    """
    try:
        # Initialize an empty list to store DataFrames
        dfs = []
        
        # Iterate over each file in the directory
        for file in os.listdir(directory):
            if file.endswith('.csv'):
                # Read the CSV file into a DataFrame
                try:
                    df = pd.read_csv(os.path.join(directory, file))
                    # Append the DataFrame to the list
                    dfs.append(df)
                except pd.errors.ParserError:
                    print(f"Error reading file: {file}")
        
        # Concatenate all DataFrames in the list vertically
        concatenated_df = pd.concat(dfs, ignore_index=True)
        
        return concatenated_df
    except FileNotFoundError:
        print("Directory not found.")
    except pd.errors.EmptyDataError:
        print("One or more files are empty.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
concatenated_sales_data = load_and_concatenate_sales_data('./sales_data')
````
